```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv("StudentsPerformance.csv")

# Preview dataset
df.head()
```

```python
# 1. Histogram → Distribution of math scores
plt.hist(df['math score'], bins=20, color='skyblue', edgecolor='black')
plt.title("Distribution of Math Scores")
plt.xlabel("Math Score")
plt.ylabel("Frequency")
plt.show()
```

```python
# 2. Line plot → Math scores trend (first 50 students)
plt.plot(df['math score'][:50], label="Math Score", marker='o', color="green")
plt.title("Math Scores Trend (First 50 Students)")
plt.xlabel("Student Index")
plt.ylabel("Score")
plt.legend()
plt.show()
```

```python
# 3. Bar chart → Average math score by gender
avg_scores = df.groupby("gender")['math score'].mean()
plt.bar(avg_scores.index, avg_scores.values, color=['blue','pink'])
plt.title("Average Math Score by Gender")
plt.ylabel("Average Score")
plt.show()
```

```python
# 4. Scatter plot → Math score vs. Reading score
plt.scatter(df['math score'], df['reading score'], alpha=0.6, c='red')
plt.title("Math vs Reading Scores")
plt.xlabel("Math Score")
plt.ylabel("Reading Score")
plt.show()
```

```python
# 5. Pie chart → Gender distribution
gender_counts = df['gender'].value_counts()
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=90)
plt.title("Gender Distribution")
plt.show()
```

```python
# 6. Stacked bar chart → Average scores (Math/Reading/Writing) by gender
avg_all = df.groupby("gender")[['math score','reading score','writing score']].mean()
avg_all.plot(kind="bar", stacked=True, figsize=(7,5))
plt.title("Average Scores by Gender (Stacked)")
plt.ylabel("Average Scores")
plt.show()
```

```python
# 1. KDE plot → Reading score distribution (updated for seaborn >=0.14)
sns.kdeplot(df['reading score'], fill=True, color="purple")
plt.title("Reading Score Distribution (KDE)")
plt.show()
```

```python
# 2 Box plot → Math score by gender (updated for seaborn >=0.14)
sns.boxplot(x="gender", y="math score", hue="gender", data=df, palette="Set2", legend=False)
plt.title("Math Scores by Gender")
plt.show()
```

```python
# 3. Violin plot → Writing score by test preparation
sns.violinplot(x="test preparation course", 
               y="writing score", 
               hue="test preparation course",  # add hue
               data=df, 
               palette="muted", 
               legend=False)  # remove duplicate legend
plt.title("Writing Scores by Test Preparation")
plt.show()
```

```python
# 4. Swarm plot → Math score by lunch type
sns.swarmplot(x="lunch", 
              y="math score", 
              hue="lunch",           # added hue
              data=df, 
              palette="Set1", 
              legend=False, 
              size=4)                # reduce point size
plt.title("Math Scores by Lunch Type")
plt.show()
```

```python
# 5. Pairplot → Relation between all three scores
sns.pairplot(df[['math score','reading score','writing score']])
plt.suptitle("Pairplot of Scores", y=1.02)
plt.show()
```

```python
# 6. Countplot → Students by parental education level
sns.countplot(x="parental level of education", 
              hue="parental level of education",   # add hue
              data=df, 
              palette="pastel", 
              legend=False)                        # avoid duplicate legend
plt.xticks(rotation=45)
plt.title("Students by Parental Education Level")
plt.show()
```

```python
# 7. Heatmap → Correlation between scores
plt.figure(figsize=(8,6))
sns.heatmap(df[['math score','reading score','writing score']].corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap of Scores")
plt.show()
```
